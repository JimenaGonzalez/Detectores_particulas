```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

<center> <h2>Partículas alfa incidiendo en diferentes medios</h2> </center>

El primer objetivo es encontrar el alcance de las partículas Alfa en diferentes medios, se escogen: Hierro, Plata y Plomo. Para esto lo primero que se hace es importar los datos de la página https://www.nist.gov/pml/stopping-power-range-tables-electrons-protons-and-helium-ions 
para iones de Helio. Los datos constan de dos columnas, en la primera se encuentra la energía de la partícula en unidades de  $MeV$ y
en la segunda el stopping power ($\frac{dE}{dx}$) en unidades de $MeV\ cm² /g$. 

```python
datos_alfa_hierro = np.genfromtxt("Alfa en Hierro.txt",delimiter="|", dtype=float)
datos_alfa_plata = np.genfromtxt("Alfa en Plata.txt",delimiter="|", dtype=float)
datos_alfa_plomo = np.genfromtxt("Alfa en Plomo.txt",delimiter="|", dtype=float)
```

Se realiza una gráfica de los datos importados para ver como se comportan.

```python
plt.loglog(datos_alfa_hierro[:,0], datos_alfa_hierro[:,1], datos_alfa_plata[:,0], datos_alfa_plata[:,1], datos_alfa_plomo[:,0], datos_alfa_plomo[:,1])
plt.title(u"Partículas alfa en diferentes medios")
plt.ylabel(r"Stopping power $(MeV cm^{2}/g)$")
plt.xlabel(u"Energía " + r"(MeV)")
plt.legend(['Hierro', 'Plata', 'Plomo'])
plt.show()
```

Dado que los datos usados es un conjuntos discreto se realiza una interpolación de los datos importados para tener una distribución continua de los mismos.

```python
afuncion_hierro = interp1d(datos_alfa_hierro[:,0], datos_alfa_hierro[:,1])
afuncion_plata = interp1d(datos_alfa_plata[:,0], datos_alfa_plata[:,1])
afuncion_plomo = interp1d(datos_alfa_plomo[:,0], datos_alfa_plomo[:,1])
```

Se definen dos variables por cada material que van a ser:
+ Variable energy_new: va a tener los valores de la energía de las partículas alfa, son valores que están limitados entre el valor mínimo de la energía en ese material y la máxima (que es la energía inicial), sus valores están igualmente separados y consta de 100 valores. Esta variable se define para cada material.
<br><br>
+ Variable stop_power: contiene los valores respectivos a las energías usando la interpolación que se calculo en el punto anterior. Esta variable se calcula para cada material.

```python
max_energia = 5
num = 100
energy_new_ahierro = np.linspace(min(datos_alfa_hierro[:,0]), max_energia, num, endpoint=True)
energy_new_aplata = np.linspace(min(datos_alfa_plata[:,0]), max_energia, num, endpoint=True)
energy_new_aplomo = np.linspace(min(datos_alfa_plomo[:,0]), max_energia, num, endpoint=True)
stop_power_ahierro = afuncion_hierro(energy_new_ahierro)
stop_power_aplata = afuncion_plata(energy_new_aplata)
stop_power_aplomo = afuncion_plomo(energy_new_aplomo)
```

+ La longitud total del material se escoge como 1 cm.
+ La longitud de cada capa de espesor se escoge como 0.0001 cm.
+ Variable energia_total: Guarda la energía de la partícula en el material, su valor inicial es de 5 $MeV$. Esta variable se define para cada material.
+ Variable lista_deltaE: Se usa para guardar los valores de stopping power de la energía correspondiente. También se define para cada material.

```python
longitud = 1
delta_x = longitud/10000.
alcance = 0
lista_alcance = []
energia_total_ahierro = 5 
energia_total_aplata = 5 
energia_total_aplomo = 5 
lista_deltaE_ahierro = []
lista_deltaE_aplata = []
lista_deltaE_aplomo = []
lista_energia_ahierro = []
lista_energia_aplata = []
lista_energia_aplomo = []
```

Los pasos que se dan en el siguiente loop son los siguientes:
+ Se calcula el delta de la energía para cada celda, esto se hace multiplicando el valor correspondiente de stopping power y multiplicandolo por la longitud dx de cada celda.
+ A la energía total se le resta el valor del delta de la energía encontrado en el paso anterior.
+ En la lista deltaE se guarda el valor correspodiente al stopping power.
+ En la lista energia se guarda el valor actual de la energía de la partícula.<br><br>
Este algoritmo se sigue para cada material. Además en la lista lista_alcance se guarda el valor correpondiente a la distancia en x, esta es una varible universal para todos los materiales.

```python
for i in range(num):
    delta_E_ahierro = stop_power_ahierro[-i-1]*delta_x
    delta_E_aplata = stop_power_aplata[-i-1]*delta_x
    delta_E_aplomo = stop_power_aplomo[-i-1]*delta_x
    energia_total_ahierro -= delta_E_ahierro
    energia_total_aplata -= delta_E_aplata
    energia_total_aplomo -= delta_E_aplomo
    lista_deltaE_ahierro.append(stop_power_ahierro[-i-1])
    lista_deltaE_aplata.append(stop_power_aplata[-i-1])
    lista_deltaE_aplomo.append(stop_power_aplomo[-i-1])
    lista_energia_ahierro.append(energia_total_ahierro)
    lista_energia_aplata.append(energia_total_aplata)
    lista_energia_aplomo.append(energia_total_aplomo)
    alcance += delta_x
    lista_alcance.append(alcance)
```

En los siguientes ciclos se calcula el alcance para cada material. Esto se hace buscando cuando
el valor de la energía es igual o menor a cero y así se encuentra el valor correspondiente a la distancia. A final se imprimen los valores de los alcances encontrados para el hierro, la plata y el plomo.

```python
for i in range(len(lista_energia_ahierro)):
    if(lista_energia_ahierro[i] <= 0):
        alcance_ahierro = lista_alcance[i]
        break
        
for i in range(len(lista_energia_aplata)):
    if(lista_energia_aplata[i] <= 0):
        alcance_aplata = lista_alcance[i]
        break
        
for i in range(len(lista_energia_aplomo)):
    if(lista_energia_aplomo[i] <= 0):
        alcance_aplomo = lista_alcance[i]
        break
        
print "Valores de los alcances R    hierro: " + str(alcance_ahierro) + "cm, plata: " + str(alcance_aplata) + "cm, plomo: " + str(alcance_aplomo) + "cm."
```

Se realiza una gráfica del stopping power vs la distancia recorrida de la partícula. En la leyenda 
se incluye el valor encontrado para el alcance en cada material.

```python
plt.plot(lista_alcance, lista_deltaE_ahierro, lista_alcance, lista_deltaE_aplata, lista_alcance, lista_deltaE_aplomo)
plt.title(u"Partículas alfa en diferentes medios")
plt.ylabel(r"Stopping power $(MeV cm^{2}/g)$")
plt.xlabel(u"Distancia " + r"(cm)")
plt.legend(['Hierro R = ' + str(alcance_ahierro) , 'Plata R = ' + str(alcance_aplata), 'Plomo R = ' + str(alcance_aplomo)])
plt.show()
```

<center> <h2>Protones y partículas alfa incidiendo en el mismo medio</h2> </center>

Para comparar el resultado obtenido para partículas Alfa con protones se escoge como medio común Hierro.
El análisis para los protones se hace de forma similar: se importan los datos de las tablas que se 
encuentran en la web, se hace una interpolación de los datos, se hace un ciclo para hallar los valores de
las variables de interés: stopping power, energia y alcance. Por último, se hace un ciclo para encontrar el 
valor del alcance para protones en Hierro (este valor se imprime en pantalla). <br><br>
A continución de realiza una gráfica de los datos importandos donde se pueden comparar los valores para los
protones y las partículas alfa en el Hierro.

```python
datos_protones_hierro = np.genfromtxt("Protones en Hierro.txt",delimiter="|", dtype=float)
plt.loglog(datos_protones_hierro[:,0],datos_protones_hierro[:,1], datos_alfa_hierro[:,0], datos_alfa_hierro[:,1])
plt.title(u"Protones y partículas alfa en Hierro")
plt.ylabel(r"Stopping power $(MeV cm^{2}/g)$")
plt.xlabel(u"Energía " + r"(MeV)")
plt.legend(['Protones', u'Partículas Alfa'])
plt.show()
```

```python
pfuncion_hierro = interp1d(datos_protones_hierro[:,0], datos_protones_hierro[:,1])
max_energia = 5
num = 100
energy_new_phierro = np.linspace(min(datos_protones_hierro[:,0]), max_energia, num, endpoint=True)
stop_power_phierro = pfuncion_hierro(energy_new_phierro)
longitud = 1
delta_x = longitud/10000.
alcance = 0
lista_alcance = []
energia_total_phierro = 5 
lista_deltaE_phierro = []
lista_energia_phierro = []
for i in range(num):
    delta_E_phierro = stop_power_phierro[-i-1]*delta_x
    energia_total_phierro -= delta_E_phierro
    lista_deltaE_phierro.append(stop_power_phierro[-i-1])
    lista_energia_phierro.append(energia_total_phierro)
    alcance += delta_x
    lista_alcance.append(alcance)
for i in range(len(lista_energia_phierro)):
    if(lista_energia_phierro[i] <= 0):
        alcance_phierro = lista_alcance[i]
        break
print "Alcance de los protones en el hierro = " + str(alcance_phierro) + "cm."
```

```python
plt.plot(lista_alcance, lista_deltaE_phierro, lista_alcance, lista_deltaE_ahierro)
plt.title(u"Protones y partículas alfa en Hierro")
plt.ylabel(r"Stopping power $(MeV cm^{2}/g)$")
plt.xlabel(u"Distancia " + r"(cm)")
plt.legend(['Protones R = ' + str(alcance_phierro), u'Partículas Alfa R= ' + str(alcance_ahierro)])
plt.show()
```
